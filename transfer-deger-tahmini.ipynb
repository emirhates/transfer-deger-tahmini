```python
# Tüm 4 dataframe için başlık temizliği
std = std[std["Player"] != "Player"]
shooting = shooting[shooting["Player"] != "Player"]
misc = misc[misc["Player"] != "Player"]
playing = playing[playing["Player"] != "Player"]

# Şimdi sayı dönüşümünü yap
std["Min"] = std["Min"].str.replace(",", "").astype(int)

# Kontrol
print(std["Min"].dtype)
print(std.shape)
```

```python
sayi_sutunlari = ["MP", "Starts", "90s", "Gls", "Ast", "PK", "PKatt", "CrdY", "CrdR"]
std[sayi_sutunlari] = std[sayi_sutunlari].apply(pd.to_numeric, errors="coerce")

print(std.dtypes)
print(std.shape)
```

```python
# SHOOTING
shooting = shooting[shooting["Player"] != "Player"]
sayi_shooting = ["Sh", "SoT", "SoT%", "Sh/90", "SoT/90", "G/Sh", "G/SoT"]
shooting[sayi_shooting] = shooting[sayi_shooting].apply(pd.to_numeric, errors="coerce")

# MISC
misc = misc[misc["Player"] != "Player"]
sayi_misc = ["Fls", "Fld", "Off", "Crs", "Int", "TklW", "2CrdY"]
misc[sayi_misc] = misc[sayi_misc].apply(pd.to_numeric, errors="coerce")

# PLAYING
playing = playing[playing["Player"] != "Player"]
sayi_playing = ["Mn/MP", "Mn/Start", "Compl", "Subs", "Mn/Sub", "unSub", "PPM"]
playing[sayi_playing] = playing[sayi_playing].apply(pd.to_numeric, errors="coerce")
```

```python
std_temiz = std[["Player", "Nation", "Pos", "Squad", "Age", "Born", "MP", "Starts", "Min", "90s", "Gls", "Ast", "PK", "PKatt", "CrdY", "CrdR"]]
shooting_temiz = shooting[["Player", "Squad", "Sh", "SoT", "SoT%", "Sh/90", "SoT/90", "G/Sh", "G/SoT"]]
misc_temiz = misc[["Player", "Squad", "Fls", "Fld", "Off", "Crs", "Int", "TklW", "2CrdY"]]
playing_temiz = playing[["Player", "Squad", "Mn/MP", "Mn/Start", "Compl", "Subs", "Mn/Sub", "unSub", "PPM"]]
```
