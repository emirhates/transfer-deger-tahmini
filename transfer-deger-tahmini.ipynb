```python
from thefuzz import process

def en_yakin_eslesme(isim, secenekler, esik=85):
    """En yakın eşleşmeyi bulur, eşik altıysa None döner"""
    sonuc = process.extractOne(isim, secenekler)
    if sonuc and sonuc[1] >= esik:
        return sonuc[0]
    return None
```

```python
fbref_liste = birlesik["Player"].tolist()
transfer["FBref_Eslesme"] = transfer["Player"].apply(lambda x: en_yakin_eslesme(x, fbref_liste))

# Kontrol
print("Eşleşen:", transfer["FBref_Eslesme"].notna().sum())
print("Eşleşmeyen:", transfer["FBref_Eslesme"].isna().sum())
```

```python
eslesmeyenler = transfer[transfer["FBref_Eslesme"].isna()]
print(eslesmeyenler[["Player", "Squad"]].head(30))
```

```python
# Eşleşen ama isimleri farklı olanları göster
farkli_eslesmeler = transfer[
    (transfer["FBref_Eslesme"].notna()) & 
    (transfer["Player"] != transfer["FBref_Eslesme"])
][["Player", "FBref_Eslesme", "Squad"]]

print("Farklı eşleşme sayısı:", len(farkli_eslesmeler))
print()
print(farkli_eslesmeler.head(30))
```

```python
def takim_bazli_eslesme(player, squad, fbref_df, esik=75):
    """Sadece aynı takımdaki oyuncularla eşleştir"""
    ayni_takim = fbref_df[fbref_df["Squad"] == squad]["Player"].tolist()
    
    if not ayni_takim:
        return None
    
    sonuc = process.extractOne(player, ayni_takim)
    if sonuc and sonuc[1] >= esik:
        return sonuc[0]
    return None

transfer["FBref_Eslesme"] = transfer.apply(
    lambda row: takim_bazli_eslesme(row["Player"], row["Squad"], birlesik), 
    axis=1
)

print("Eşleşen:", transfer["FBref_Eslesme"].notna().sum())
print("Eşleşmeyen:", transfer["FBref_Eslesme"].isna().sum())
```

```python
farkli_eslesmeler = transfer[
    (transfer["FBref_Eslesme"].notna()) & 
    (transfer["Player"] != transfer["FBref_Eslesme"])
][["Player", "FBref_Eslesme", "Squad"]]

print("Farklı eşleşme sayısı:", len(farkli_eslesmeler))
print()
print(farkli_eslesmeler.head(40))
```
